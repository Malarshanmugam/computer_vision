```python
%matplotlib notebook
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_1 = 'images/ball.jpg'
file_2 = 'images/bulb.jpg'
```

```python
cv_img_1 = cv2.imread(file_1)
```

```python
cv_img_1 = cv2.cvtColor(cv_img_1,cv2.COLOR_BGR2RGB)
```

```python
cv_img_1 = cv_img_1.astype('float32')
```

```python
hsv_cv_img = cv2.cvtColor(cmy_cv_img,cv2.COLOR_RGB2HSV)
```

```python
hsv_cv_img
```

```python
plt.imshow(hsv_cv_img[:,:,1],cmap='hsv')
```

```python
lwr = np.array([220,0.5,0.5])
hir = np.array([240,1.0,1.0])
```

```python
mask = cv2.inRange(hsv_cv_img,lwr,hir)
```

```python
plt.imshow(mask,cmap='gray')
```

```python
comb = cv2.bitwise_and(cv_img_1,cv_img_1,mask=mask)
```

```python
plt.imshow(comb/255)
```

```python
plt.imshow(mask,cmap='gray')
```

```python
conts = cv2.findContours(mask,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
```

```python
conts
```

```python
conts[0]
```

```python
cnt = conts[0]
```

```python
M = cv2.moments(cnt)
M
```

```python
cx = int(M['m10']/M['m00'])
cy = int(M['m01']/M['m00'])
```

```python
print(cx)
print(cy)
```
