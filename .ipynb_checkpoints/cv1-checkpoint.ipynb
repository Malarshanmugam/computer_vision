```python
image = cv2.imread('images/cv_im.jpg')
cv2.imshow('tab',image)
cv2.waitKey()
cv2.destroyWindow('tab')
```

```python
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
plt.figure()
plt.imshow(gray)
plt.colorbar()
```

```python
plt.figure()
plt.imshow(gray, cmap='gray')
plt.colorbar()
```

```python
gray_fl = gray.astype('float64')/255
print(gray_fl)
print(gray_fl.dtype)
```

```python
ht, wt = gray_fl.shape
print(ht,"",wt)
```

```python
rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure()
plt.imshow(rgb)
```

```python
rgb_r = rgb[:,:,0]
rgb_g = rgb[:,:,1]
rgb_b = rgb[:,:,2]
```

```python
plt.figure()
plt.imshow(rgb_r)
plt.colorbar()
```

```python
plt.figure()
plt.imshow(rgb_g)
```

```python
plt.figure()
plt.imshow(rgb_b)
```

```python
print(rgb_r.shape)
```

```python
combined = np.zeros((ht, wt, 3), 'uint8')
combined[:,:,0] = rgb_r
combined[:,:,1] = rgb_g
combined[:,:,2] = rgb_b
plt.figure()
plt.imshow(combined)
```

```python
air_b1 = combined[0:50, 90:150]
plt.figure()
plt.imshow(air_b1)
```

```python
combined[0:50, 90:150] = 0
plt.figure()
plt.imshow(combined)
```

```python
combined[0:50,0:50,0] = 0
plt.figure()
plt.imshow(combined)
```
