## Image Filters

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/yell2.jpg')
```

```python
# custom kernel filtering
kernel = np.ones((5,5),np.float32)
kernel /= kernel.size
blur = cv2.filter2D(img, -1, kernel)
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.imshow(img)
plt.title('original')
plt.subplot(1,2,2)
plt.imshow(blur)
plt.title('blur')
```

```python
# average blurring
blur2 = cv2.blur(img, (5,5))
plt.figure(figsize=(10,10))
plt.imshow(blur2)
```

```python
# average blur
blur3 = cv2.boxFilter(img, -1, (5,5))
plt.figure()
plt.imshow(blur3)
```

```python
# gaussian blur
blur4 = cv2.GaussianBlur(img, (11,11), 0)
plt.figure()
plt.imshow(blur4)
```

```python
# median blur
blur5 = cv2.medianBlur(img, 5)
plt.figure()
plt.imshow(blur5)
```

```python
sap = cv2.imread('images/salt and pepper.jpg')
plt.imshow(sap, cmap='gray')
```

```python
sap_median = cv2.medianBlur(sap, 19)
plt.imshow(sap_median, cmap='gray')
```

```python
sap_gaussian = cv2.GaussianBlur(sap, (19, 19), 0)
plt.imshow(sap_gaussian,  cmap='gray')
```

```python
gn = cv2.imread('images/gaussian.jpg')
plt.imshow(gn, cmap='gray')
```

```python
gn_gaussian = cv2.GaussianBlur(gn, (19,19),0)
plt.imshow(gn_gaussian)
```

```python
gn_median = cv2.medianBlur(gn, 19)
plt.imshow(gn_median, cmap='gray')
```

## Sharpening of images

```python
sap_lap = cv2.Laplacian(sap, cv2.CV_64F)
plt.imshow(sap_lap)
```

```python
sudoku = cv2.imread('images/sudoku_2.jpg', 0)
plt.imshow(sudoku, cmap='gray')
```

```python
sudoku_lap = cv2.Laplacian(sudoku, cv2.CV_64F)
plt.imshow(sudoku_lap, cmap='gray')
```

```python
sudoku_sharp = sudoku - sudoku_lap
```

```python
plt.imshow(sudoku_sharp, cmap='gray')
```

```python
sobelx = cv2.Sobel(sudoku, cv2.CV_64F, 1, 0, ksize=5)
plt.imshow(sobelx, cmap='gray')
```

```python
sobely = cv2.Sobel(sudoku, cv2.CV_64F, 0, 1, ksize=5)
plt.imshow(sobely, cmap='gray')
```

```python
sudoku_gauss = cv2.GaussianBlur(sudoku,(19,19),0)
plt.imshow(sudoku_gauss, cmap='gray')
```

```python
mask = sudoku - sudoku_gauss
```

```python
imp = sudoku + mask
```

```python
plt.imshow(imp, cmap='gray')
```
